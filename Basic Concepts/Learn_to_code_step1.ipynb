### 7.

Now, can you calculate the simple interest for a 3 year, 5 year, and a 7 year period?

Did you write the simple interest formula thrice? Once each for 3, 5, and 7 years?

How about a more convenient way? Here's how you can make it easier - 

Calculate the simple interest for one year and store it in a `variable` (a variable is like a box that you use for storing values so you can use them later).

```python
y = 500*10/100
```

You can now use this pre-calculated value to find the interest amount for any duration - 

```python
# 3 years

3*y
```

```python
# 5 years

5*y
```

```python
# 7 years

7*y
```

```python
# Maybe... 32 years!

32*y
```
